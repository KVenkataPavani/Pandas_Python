#**Creating a DataFrame in Pandas and Merge Concate Join**


```python
import pandas as pd
```

Creating a DataFrame in Pandas


```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)
df
```

```python
# From lists of lists:
data = [['Alice', 25, 'New York'], ['Bob', 30, 'Los Angeles'], ['Charlie', 35, 'Chicago']]
df = pd.DataFrame(data, columns=['Name', 'Age', 'City'])
df
```

1. merge() in Pandas


```python
#  merge() in Pandas
# pd.merge(left, right, on='column_name', how='join_type')
# Parameters:
# left: The first DataFrame.
# right: The second DataFrame.
# on: The common column to merge on (use this if the column name is the same in both DataFrames).
# how: Specifies the type of join to perform. Options include:
# 'inner': Returns rows with matching values in both DataFrames (default).
# 'outer': Returns all rows from both DataFrames, filling in NaN for missing values.
# 'left': Returns all rows from the left DataFrame and matching rows from the right.
# 'right': Returns all rows from the right DataFrame and matching rows from the left.

df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [30, 40, 50]})

merged_df1 = pd.merge(df1, df2)
print(merged_df1)

# Inner join on 'ID'
merged_df = pd.merge(df1, df2, on='ID', how='outer')
merged_df
```

2. concat() in Pandas


```python
# . concat() in Pandas
# The concat() function in Pandas is used to concatenate or stack multiple DataFrames along a particular axis.
# pd.concat([df1, df2], axis=0, join='outer')

# Parameters:
# objs: List of DataFrames to concatenate.
# axis: The axis to concatenate along. axis=0 (default) stacks rows vertically, and axis=1 stacks columns horizontally.
# join: Defines how to handle indexes. Options include:
# 'outer': Includes all indexes (union).
# 'inner': Includes only common indexes (intersection)
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'], 'B': ['B0', 'B1', 'B2']})
df2 = pd.DataFrame({'A': ['A3', 'A4'], 'B': ['B3', 'B4']})

# Concatenate rows (default)
concatenated_df = pd.concat([df1, df2])
concatenated_df
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2']})
df2 = pd.DataFrame({'B': ['B0', 'B1', 'B2']})

# Concatenate columns (axis=1)
concatenated_df = pd.concat([df1, df2], axis=1)
concatenated_df
```


3. join() in Pandas

```python
# join() in Pandas
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2']}, index=['K0', 'K1', 'K2'])
df2 = pd.DataFrame({'B': ['B0', 'B1', 'B2']}, index=['K0', 'K1', 'K3'])

# Left join on index
joined_df = df1.join(df2, how='left')
joined_df
```
