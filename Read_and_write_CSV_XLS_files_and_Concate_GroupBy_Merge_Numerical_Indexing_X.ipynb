# **concatenate Data Frames**

```python
import pandas as pd
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})

india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
#concate two dataframes
df = pd.concat([india_weather, us_weather])
df
```

```python
#if you want continuous index
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
df = pd.concat([india_weather, us_weather],axis=1)
df
#shows in rows not like column line downward
```

#**Merge DataFrames**

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore", 'hyderabad'],
    "temperature": [32,45,30,40]})
temperature_df
```

```python
humidity_df = pd.DataFrame({
    "city": ["delhi","mumbai","banglore"],
    "humidity": [68, 65, 75]})
humidity_df
```

```python
#merge two dataframes with out explicitly mention index
df = pd.merge(temperature_df, humidity_df, on='city')
df
#in merge hyd drops bcz theres no humidity
```

```python
#OUTER-JOIN
df = pd.merge(temperature_df, humidity_df, on='city', how='outer')
df
#in outer join hyd will also be printed even another value is not there also it shows NaN not a number
```

# **Numerical Indexing (.loc vs iloc)**

```python
import pandas as pd
import numpy as np
```

```python
#row n is equal to default index i.e starts from 0,1,2
df = pd.DataFrame([1,2,3,4,5,6,7,8,9,19])#, index=[49,48,47,46,45, 1, 2, 3, 4, 5])
df
```

```python
#row n is equal to default index i.e starts from 0,1,2
df = pd.DataFrame([1,2,3,4,5,6,7,8,9,19], index=[49,48,47,46,45, 1, 2, 3, 4, 5]) #to give index of our choice
df
```

```python
print(df.loc[1])#loc[index]
#go to that index value 1 get me the value stored
print(df.loc[48])
print(df.loc[47])
print(df.loc[5])

```

```python
print(df.iloc[1])#loc[row number]
#go to the index value 1 get me the value stored
print(df.iloc[0])
print(df.iloc[5])
print(df.iloc[45])
```

```python
df.loc[:2]
#includes 2 index value also
```

```python
df.iloc[:2]
#index location  1st 2 rows we get
```
